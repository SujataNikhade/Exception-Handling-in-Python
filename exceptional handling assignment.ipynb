```python
#Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.
try:
    n=int(input("enter any num: "))
    print(n/0)
except:
    print("you can't divide any num by zero")
```

```python
#prompts the user to input an integer and raises a ValueError exception if the input is not a valid integer.
try:
    a=int(input("enter the number: "))
    print (a)
except:
    print("Invalid input! please enter the valid number")
```

```python
#Write a Python program that opens a file and handles a FileNotFoundError exception if the file does not exist.
try:
    n=open("sujju.txt","r")
    print(n.read())
except:
    print("the file does not exist")
```

```python
#Write a Python program that prompts the user to input two numbers
#and raises a TypeError exception if the inputs are not numerical.
try:
    a=eval(input("enter the first num: "))
    b=eval(input("enter the second num: "))
    print(a+b)
except:
    print("Typeerror")
```

```python
#Write a Python program that opens a file and handles a PermissionError exception if there is a permission issue.
try:
    f = open("C:/Windows/system.ini", "w")
    print("hello")
except:
    print("Permission error")
```

```python
##Write a Python program that executes an operation on a list and
#handles an IndexError exception if the index is out of range
try:
    lst=[10,20,30]
    index=int(input("enter the index: "))
    print(lst[index])
except:
    print("Index is out of range")
```

```python
##Write a Python program that executes division and handles an
##ArithmeticError exception if there is an arithmetic error.
try:
    a=int(input("enter the first num: "))
    b=int(input("enter the second num: "))
    print(a/b)
except:
    print("Arithmetric error do not enter the zero")
finally:
    print("Operation finished")
```
